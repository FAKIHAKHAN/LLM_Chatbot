```python
def process_pdf(pdf_file):
    """Loads PDF, extracts text, formats it with LLM, chunks, embeds, stores in vector DB."""
    if pdf_file is None:
        return "❗ Please upload a PDF first."

    # Save uploaded file
    with tempfile.NamedTemporaryFile(delete=False, suffix=".pdf") as tmp:
        tmp.write(pdf_file.read())
        pdf_path = tmp.name

    # --- Load PDF ---
    loader = PyPDFLoader(pdf_path)
    pages = loader.load()

    # Merge pages
    raw_text = "".join([p.page_content for p in pages])

    # --- Markdown splitting ---
    md_splitter = MarkdownHeaderTextSplitter(
        headers_to_split_on=[("#", "Section"), ("##", "Lecture")]
    )
    md_docs = md_splitter.split_text(raw_text)

    # --- Format text with LLM ---
    formatting_prompt = ChatPromptTemplate.from_messages([
        ("system", """
        Format the following Tableau transcript by:
        - Creating structured paragraphs
        - Fixing punctuation
        - Correcting transcription errors
        - Keeping meaning intact
        """),
        ("human", "{lecture}")
    ])

    formatting_chain = formatting_prompt | LLM | StrOutputParser()
    formatted_list = formatting_chain.batch(
        [d.page_content for d in md_docs]
    )

    # Replace contents
    for doc, new_text in zip(md_docs, formatted_list):
        doc.page_content = new_text

    # --- Token Chunking ---
    splitter = TokenTextSplitter(
        encoding_name="cl100k_base",
        chunk_size=500,
        chunk_overlap=50
    )
    chunks = splitter.split_documents(md_docs)

    # --- Vector DB ---
    vectordb = Chroma.from_documents(
        documents=chunks,
        embedding=EMBEDDINGS,
        persist_directory=VDB_DIR
    )

    return "✅ PDF successfully processed! You can now chat with your AI tutor."


def chat_with_bot(message, history):
    """Answers a query using RAG over the vector DB."""
    try:
        vectordb = Chroma(
            persist_directory=VDB_DIR,
            embedding_function=EMBEDDINGS
        )
        retriever = vectordb.as_retriever(
            search_type="mmr",
            search_kwargs={"k": 3, "lambda_mult": 0.7}
        )

        docs = retriever.get_relevant_documents(message)
        context = "\n\n".join([d.page_content for d in docs])

        prompt = ChatPromptTemplate.from_messages([
            ("system",
             "You are an AI tutor specialized in Tableau. Answer using the lecture notes."),
            ("human", "Question:\n{question}\n\nContext:\n{context}")
        ])

        answer_chain = prompt | LLM | StrOutputParser()
        answer = answer_chain.invoke({"question": message, "context": context})

        history.append((message, answer))
        return history, history

    except Exception as e:
        return history + [(message, f"❗ Error: {str(e)}")], history
```

```python
# ====== GRADIO UI ======
with gr.Blocks(theme=gr.themes.Soft(), css="footer {visibility: hidden}") as demo:

    gr.Markdown("""
    # 📊 Tableau AI Tutor
    Upload a lecture PDF and chat with a customized AI tutor trained on your content.
    """)

    with gr.Tab("📄 Upload & Process PDF"):
        with gr.Row():
            pdf_upload = gr.File(label="Upload your Tableau PDF", file_types=[".pdf"])
        process_btn = gr.Button("Process PDF")
        process_output = gr.Markdown()

        process_btn.click(
            process_pdf,
            inputs=pdf_upload,
            outputs=process_output
        )

    with gr.Tab("💬 Chat with Tutor"):
        chatbot = gr.Chatbot(
            height=450,
            label="Ask anything about your lecture"
        )
        msg = gr.Textbox(
            placeholder="Ask a question about Tableau...",
            label="Your Question"
        )
        send_btn = gr.Button("Send")

        send_btn.click(
            chat_with_bot,
            inputs=[msg, chatbot],
            outputs=[chatbot, chatbot]
        )
        msg.submit(
            chat_with_bot,
            inputs=[msg, chatbot],
            outputs=[chatbot, chatbot]
        )

demo.launch()
```
